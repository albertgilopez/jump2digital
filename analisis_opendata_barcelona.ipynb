```python
# Importación de paquetes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config IPCompleter.greedy=True 
```

```python
# ENTENDER LOS FICHEROS

# En la web de Open Data del Ayuntamiento de Barcelona podemos descargar los datasets:
# Por ejemplo, el del precio del alquiler: https://opendata-ajuntament.barcelona.cat/data/es/dataset/est-mercat-immobiliari-lloguer-mitja-mensual/resource/0a71a12d-55fa-4a76-b816-4ee55f84d327 

df_alquiler = pd.read_csv('2017_lloguer_preu_trim.csv')

df_alquiler.info()
df_alquiler.head()
```

```python
# Hacemos lo mismo con el cojunto de datos de exposición al ruído

df_ruido = pd.read_csv('2017_poblacio_exposada_barris_mapa_estrategic_soroll_bcn_long.csv')

df_ruido.info()
df_ruido.head()
```

```python
# Ambos conjuntos de datos tienen columnas para el código y el nombre del distrito y del barrio, lo que facilita la unión de los datos.

df_unificado = pd.merge(df_alquiler, df_ruido, on=['Codi_Districte', 'Nom_Districte', 'Codi_Barri', 'Nom_Barri'], how='inner')

df_unificado.info()
df_unificado.head()
```

```python
columnas = ['Valor']

for col in columnas:
    valores_unicos = df_unificado[col].unique()
    print(f"Valores únicos para {col}:")
    print(valores_unicos)
    print("\n")
```

Se han unificado los conjuntos de datos de alquileres y exposición al ruido. Ahora tenemos un solo conjunto de datos que contiene información sobre el precio medio de alquiler en diferentes barrios de Barcelona, así como el porcentaje de la población expuesta a diferentes niveles de ruido en esos mismos barrios.

```python
# Examinar las filas con valores nulos en la columna 'Preu' para entender el patrón, si lo hay
df_nulos = df_unificado[df_unificado['Preu'].isnull()]

# Mostrar algunas estadísticas y muestras de los datos con valores nulos
estadisticas_nulos = df_nulos.describe(include='all')
muestras_nulos = df_nulos.sample(10) 

estadisticas_nulos, muestras_nulos
```

Dado que los valores nulos parecen concentrarse en ciertas áreas, podríamos investigar si estos barrios o distritos tienen alguna característica particular que justifique la falta de datos en el precio del alquiler. Otra opción sería eliminar estas filas si consideramos que su falta de información sobre el precio del alquiler las hace menos útiles para nuestro análisis.

```python
barrios_nulos = df_nulos['Nom_Barri'].value_counts()
distritos_nulos = df_nulos['Nom_Districte'].value_counts()

barrios_nulos, distritos_nulos
```

Vamos a seguir investigando un poco más:

```python
# Investigar más a fondo los niveles de ruido en los barrios con valores nulos en la columna 'Preu'
ruido_barrios_nulos = df_nulos.groupby('Nom_Barri')['Rang_soroll'].value_counts().unstack().fillna(0)

# Investigar más a fondo los niveles de ruido en los distritos con valores nulos en la columna 'Preu'
ruido_distritos_nulos = df_nulos.groupby('Nom_Districte')['Rang_soroll'].value_counts().unstack().fillna(0)

ruido_barrios_nulos, ruido_distritos_nulos
```

* Barrios con Valores Nulos: Todos los barrios con valores nulos en el precio de alquiler (Baró de Viver, Can Peguera, Torre Baró, Vallbona, la Clota, la Marina del Prat Vermell) tienen registros para todos los rangos de niveles de ruido. Esto sugiere que la exposición al ruido está bien documentada en estos barrios.

* Distritos con Valores Nulos: De manera similar, los distritos con valores nulos en el precio de alquiler (Horta-Guinardó, Nou Barris, Sant Andreu, Sants-Montjuïc) también tienen registros para todos los rangos de niveles de ruido.

Esto podría indicar que aunque no tenemos información sobre los precios de alquiler en estos barrios y distritos, sí tenemos datos sobre la exposición al ruido. La falta de datos de alquiler podría deberse a una variedad de razones, como una baja actividad de alquiler o simplemente una falta de recopilación de datos, como comentabamos antes.

Una idea sería mirar precios de barrios adyacente para saber si podemos imputar algun pecio medio.

Sin embargo, dado que no tenemos acceso a datos geográficos adicionales, no podemos determinarlo directamente. Una opción podría ser asumir que los barrios dentro del mismo distrito podrían tener características similares y, por lo tanto, podríamos considerar imputar la media o la mediana de los precios de alquiler del distrito correspondiente.



```python
# Calcular la media del precio de alquiler por distrito, excluyendo los valores nulos
media_por_distrito = df_unificado[df_unificado['Preu'].notna()].groupby('Nom_Districte')['Preu'].mean()

# Imputar los valores nulos en la columna 'Preu' con la media del precio de alquiler del distrito correspondiente
df_unificado['Preu'].fillna(df_unificado['Nom_Districte'].map(media_por_distrito), inplace=True)

# Verificar si aún hay valores nulos en la columna 'Preu'
valores_nulos_post_imputacion = df_unificado['Preu'].isnull().sum()

valores_nulos_post_imputacion, df_unificado.head()
```

```python
# Quitar el símbolo '%' de la columna 'Valor' y convertirla a float
df_unificado['Valor'] = df_unificado['Valor'].str.rstrip('%').astype('float')

# Convertir las columnas categóricas a tipo 'category' para optimizar memoria
columnas_categoricas = ['Nom_Districte', 'Nom_Barri', 'Rang_soroll']
df_unificado[columnas_categoricas] = df_unificado[columnas_categoricas].astype('category')
```

```python
# Función para segmentar los rangos de ruido en categorías más generales
# https://www.rtve.es/noticias/20100328/niveles-de-decibelios-db-en-nuestro-entorno/322078.shtml

def segmentar_ruido(rango):
    if rango in ["< 40 dB", "40-45 dB", "45-50 dB"]:
        return "Bajo"
    elif rango in ["50-55 dB", "55-60 dB", "60-65 dB", "65-70 dB", "70-75 dB"]:
        return "Moderado"
    else:
        return "Alto"
    
# Aplicar la función de segmentación a la columna 'Rang_soroll'
df_unificado['Rang_soroll_segmentado'] = df_unificado['Rang_soroll'].apply(segmentar_ruido)

# Convertir la nueva columna a tipo 'category'
df_unificado['Rang_soroll_segmentado'] = df_unificado['Rang_soroll_segmentado'].astype('category')

# Mostrar las primeras filas del DataFrame para verificar los cambios
df_unificado.head()
```
